```python
#Reason column : ems, fire,traffic
df['reason']=df['title'].apply(lambda title: title.split(':')[0])
```

```python
df['reason'].value_counts()
```

```python
sns.countplot(x='reason',data=df,palette='viridis')
```

```python
type(df['timeStamp'].iloc[0])
```

```python
df['timeStamp'].iloc[0]
```

```python
df['timestamp']=pd.to_datetime(df['timeStamp'])
```

```python
time=df['timestamp'].iloc[0]
time.hour
```

```python
df['hour']=df['timestamp'].apply(lambda time: time.hour)
df['month']=df['timestamp'].apply(lambda time: time.month)
df['day of week']=df['timestamp'].apply(lambda time: time.dayofweek)
```

```python
df['day of week']
```

```python
dmap={0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'sun'}
```

```python
dmap
```

```python
df['day of week']=df['day of week'].map(dmap)
df['day of week']
```

```python
sns.countplot(x='day of week', hue='reason',data=df)
plt.legend(loc=2)
```

```python
sns.countplot(x='month', hue='reason',data=df)
plt.legend(loc=2)
```

```python
aylar_uzre=df.groupby('month').count()
aylar_uzre
```

```python
aylar_uzre['twp'].plot()
```

```python
sns.lmplot(x='month',y='twp',data=aylar_uzre.reset_index())
```

```python
df['date']=df['timestamp'].apply(lambda time: time.date())
df['date']
```

```python
df.groupby('date').count()['twp'].plot()
plt.tight_layout()
```

```python
df[df['reason']=='Traffic'].groupby('date').count()['twp'].plot()
plt.tight_layout()
plt.title('traffic')
```

```python
df[df['reason']=='EMS'].groupby('date').count()['twp'].plot()
plt.tight_layout()
plt.title('EMS')
```

```python
df[df['reason']=='Fire'].groupby('date').count()['twp'].plot()
plt.tight_layout()
plt.title('Fire')
```

```python
gunsaat=df.groupby(by=['day of week','hour']).count()['reason'].unstack()
gunsaat
#
```

```python
plt.figure(figsize=(18,8))
sns.heatmap(gunsaat,cmap='viridis')
```

```python
gunay=df.groupby(by=['day of week','month']).count()['reason'].unstack()
gunay
```

```python
plt.figure(figsize=(18,8))
sns.heatmap(gunay,cmap='viridis')
#
```
